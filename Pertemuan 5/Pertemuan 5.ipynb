1. Visualize top 10 teams points that have the most wins in their Home games
2. Visualize La Lakers (LAL) number of wins per year starting from 2010

```python
df[
    (df["game_location"]=="H") &
    (df["game_result"]=="W")
].groupby("team_id")["pts"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
plt.show()
```

```python
df[
    (df["team_id"]=="LAL") &
    (df["year_id"]>=2010) &
    (df["game_result"]=="W")
].groupby("year_id")["game_result"].count().plot(kind="bar")

plt.show()
```
